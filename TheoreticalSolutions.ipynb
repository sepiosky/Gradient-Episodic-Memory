# سوال ۴



روش های معمول در یادگیری مستمر برای جلوگیری از کاهش عملکرد یادگیرنده در زمان مواجهه با وظایف جدید که با وظایف قبلی یادگرفته شده تفاوت معنایی زیادی دارند از روش های مختلفی استفاده میکنند. روش هایی که سعی در منظم سازی تغییرات پارامتر های شبکه در مواجهه با تسک جدید دارند تا از تغییرات شدید مدل و کاهش عملکرد احتمالی روی تسک های قبلیی جلوگیری کنند در مواجهه با تسک های بسیار متفاوت قاعدتا در یادگیری تسک جدید انعطاف پذیریی لازم را ندارند ضعیف عمل خواهند کرد. هچنین روش هایی که کورکورانه و بدون توجه به سختی و ارتباط تسک جدید با تسک های قبلی برای هر تسک جدید تعدادی پارامتر ااضافه میکنند نیز اپتیمال نیستند زیرا هم از پتانسیل پارامتر های یادگرفته شده در تسک های قبلی استفاده مجدد کامل نمیکنند و هم اضافه کردن پارامتر هایشان ممکن ااست واقعا بیش از حد لازم باشد و مقیاس پذیری مدل را کاهش دهد.

روش این مقاله این است که ابتدا قسمت هایی از شبکه تا به اینجا آموزش داده شده که به کارایی تسک جدید مرتبط است شناسایی شوند. سپس این زیر مجموعه از پارامتر ها روی تسک جدید تقویت شوند (و احتمالا عملکرد تسک های قبلی را نیز کاهش نخواهند داد) سپس اگر همچنان در تسک جدید ضعیف بود یعنی تسک تفاوت های بیشتری با تسک های گذشته دارد و احتمالا نیاز به پارامتر هایی مختص تسک داریم. با روشی بهینه تعداد کمی پارامتر برای تسک جدید اضافه می شود. همچنین اگر مشاهده شد که تسک جدید تفاوت اساسی با تسک های قبلی داشت و حتی پیدا کردن یک زیر شبکه مرتبط با تسک سخت است و یا ادامه آموزش آن زیر شبکه عملکرد تسک های قبلی را ضعیف میکند و همچنین اضافه کردن پارامتر های جدید نیز به دلیل متصل شدن به بعضی پارامتر های قدیم باعث فراموشی میشوند مکانیزمی برای در بر گرفتن این ۲ فضای وظیفه متفاوت استفاده می شود که می تواند کپی کردن یک شبکه جدید و آموزش مجزای آن باشد.

مرحله اول که **selective retraining** نام دارد به منظور کاهش محاسبات آموزش دوباره شبکه و همچنین جلوگیری از تغییر پارامتر های نامرتبط به تسک جدید و کاهش عملکرد مدل روی تسک های قبلی به دلیل تغییر این پارامتر ها، به دنبال زیر شبکه ای کوچکتر و مرتبط با تسک جدید می گردد. به این منظور اولا شبکه روی تسک اول با یک منظم ساز L1 آموزش داده می شود تا کانکشن های نورون ها پراکنده باشد و مولفه های مختلفی در شبکه ایجاد شود، سپس روی تسک جدید پارامتر های شبکه تا قبل از لایه آخر جدا و فریز میشوند، یک لایه خطی روی آنها آنها برای دیتای تسک جدید آموزش داده میشود (با منظم ساز L1). سپس کانکشن های پراکنده ای که در لایه آخر بدست می آیند، نورون هایی از لایه ماقبل خرجی مدل تسک های قبلی که به عملکرد تسک جدید کمک میکنند را شناسایی میکند، سپس با BFS مولفه های گراف شبکه  که به این نورون ها مسیر دارند را در شبکه اسپارس قبلی پیدا میکند. در آخر کل این زیر شبکه پیدا شده روی داده های تسک جدید آموزش داده می شود. البته زیر شبکه جدید دیگر لازم نیست با L1 آموزش داده شود زیرا دیگر هدف اسپارس بودن نیست.

بعد از انجام این کار اگر تسک جدید از لحاظ معنایی به تسک های قبلی مشابه باشد احتمالا پارامتر های پیدا شده فیچر های لازم را پوشش خواهند داد و همچنین روی تسک های قبلی نیز کاهش عملکرد نخواهند داشت. اما اگر این پارامتر ها کافی نباشند یعنی تسک به فیچر های متفاوتی نیز نیاز دارد. به این منظور در فاز **Dynamic Network Expansion** سعی در اضافه کردن تعداد کافی و تا حد امکان کمی برای نیاز های تسک جدید است. این کار زمانی صورت می گیرد که مقداار تابع هزینه تسک جدید بعد از بازآموزش زیرشبکه مرتبط همچنان از حدی بالاتر باشد. در این حالت شبکه قبلی فریز می شود و سپس به هر لایه k نورون اضافه می شود. اما در بهینه سازی نورون های جدید هم منظم سازی L1 در سطح وزن هاانجام می شود تا شبکه اسپارس باقی بماند و هم روی ورودی های هر نورون منظم سازی L2 در سطح نورون ها انجام می شود. این تکنیک که group sparse regularization نام دارد علاوه بر اسپارس بودن کانکشن ها و ماتریس وزن ها، با مقید کردن بهینه سازی در سطح نورون ها باعث می شود که وزن های ورودی و خروجی نورون ها یا هم زمان صفر باشند یا همگی غیر صفر. در این صورت علاوه بر اسپارس بودن، باعث ایجاد گروه بندی بیین نورون ها و فشرده سازی بهتر قسمت های مختلف خواهیم شد. با انجام این منظم سازی روی نورون های جدید، بعد از آموزش روی تسک جدید می توان نورون هایی که کانکشن ندارند و در یادگیری بی تاثییرر هستند را حذف کرد تا فقط تعداد لازمی پارامتر به شبکه اضافه شود و از اطلاعات شبکه قبلی به خوبی استفاده مجدد صورت گیرد.


همانطور که گفته شد روش های معمول برای جلوگیری از فراموشی که به دلیل وجود تفاوت زیاد در ماهیت تسک های جدید و شیفت شبکه به یادگیری آنها و کاهش عملکرد روی تسک های قبلی رخ می دهد، نوعی مقید سازی روی آپدیت وزن ها انجام می دهند که تغییرات زیادی شاهد نباشیم تا عملکرد تسک های قبلی حفظ شود. اما ممکن است تسک ها واقعا انقدر متفاوت باشند که نیاز به تغییر داشته باشیم و با این مقید سازی عملکرد روی تسک های جدید خوب نباشد. بدین منظور این مقاله روشی برای تشخیص زمانی ارائه می دهد که ممکن است فراموشی رخ دهد و در آنجا پارامتر های مشکل ساز را با الگوریتم **Network Split/Duplication.** شناسایی و کپی میکند تا عملکرد تسک های قبلی نیز حفظ شود. در این حالت بعد از اینکه دو مرحله قبل انجام شد دوباره یک بهینه سازی روی کل شبکه (شبکه اصلی، زیری شبکه های مرتبط پیدا شده که یکبار تنظیم شده بودند و نورون های اضافه شده جدید) انجام می شود. این بهینه سازی مانند روش های معمول یادگیریی مستمر با یک قید منظم سازی روی مقدار تغییر وزن های شبکه انجام می شود. قسمت خوبی از این شبکه برای تسک جدید تنظیم شده است و کانورج کردن آن احتمالا سریع است. حال  نورون هایی که وزن های ورودی آنها از حد بیشتری تغییر کرده باشند احتمالا خروجی آن فیچر ها در اداپتیش به تسک جدید تغییر معنا دار داشته است و باعث فراموشی در تسک های قبلی خواهد شد. به این منظور این نورون ها را کپی کرده و نورون های جدید را نیز با کپی وزن های ورودی به نورون اصلی در شبکه قرار می دهیم. سپس دوباره آموزش را انجام می دهیم و انتظار داریم نورون های اولیه روی تسک های قبلی خوب عمل کنند و کپی ها با تغییر به تسک جدید اداپت شوند. آموزش دوباره شبکه با نورون های کپی شده نیز به همان دلیل احتمالا سریع کانورج خواهد کرد.

طبق آزمایش های این مقاله لایه های عمیق تر و نزدیک به خروجی در فراموشی تاثیر گذارند در حالی که لایه ها اولیه ی شبکه در طی یادگیریی مستمر پایدار تر هستند. برای بررسی فراموشی فاجعه بار این مقاله یادگیری مستمر با ۲ تسک را بررسی می کند و ابتدا شبکه را روی تسک اول آموزش داده و سپس برای آموزش روی تسک دوم شبکه از ابتدایی آن تا یک لایه L در وسط شبکه فریز می شود و آموزش روی پارامتر های لایه های بعد از L انجام می شود. در این آزمایش برای مقادیر مختلف L دو متریک مختلف بررسی می شود که در ادامه توضیح داده می شود. با توجه به نتایج لایه های ابتدایی به آسانی می توانند فریز شوند بدون اینکه برای یادگیری تسک دوم مشکل به وجود آید. این امر نشان دهنده این است فیچر های لایه های ابتدایی فیچر های جنرال قابل استفاده در تسک های دیگر هستند در حالی که هرر چه جلوتر می رویم فیچر ها بیشتر مختص تسک می شوند و عملکرد پایین روی تسک دوم وقتی آنها فریز هستند نشان دهنده همین موضوع هست. همچنین هنگامی که این لایه ها فریز نیستند در آزمایش ها میبینیم که عملکرد تسک اول بعد از آموش تسک دوم کاهش می یابد که نشان میدهد لایه های عمیق تر که به تسک ۲ اداپت می شوند مسئول فراموشی تسک های قبلی هستند.

معیار اولی که استفاده می شود CKA است که میزان شباهت بازنمایی اکتیویشن های خروجی ۲ لایه مخفی مختلف را روی یک مجموعه داده در شبکه های عصبی بررسی میکند. این معیار برای لایه های مختلف شبکه پیش از آموزش روی تسک دوم (با وزن های بدست آمده از تسک اول) و پس از آموزش مدل روی تسک دوم بررسی می شود. در این آزمایش این معیار برای لایه های اولیه مقدار بالایی دارد که نشان دهنده این است که فیچر های لایه های پایین در تسک های بعدی نیز قابل استفاده مجدد هستند و آموزش روی داده های تسک های بعدی آنها را تغییر زیادی نمی دهد. اما این مقدار برای لایه های عمیق تر کمتر است که حاکی از تغییر بازنمایی نورون ها بعد از آموزش در تسک های بعدی و غیر قابل استفاده بودن فیچر های آن ها در تسک های مختلف است.

معیار دوم برای بررسی دقیق تر این تغییر فضای بازنمایی است. در این معیار تعدادی از مولفه های اساسی فضای بازنمایی اکتیویشن های ۲ لایه مختلف با PCA بدست می آید و با محاسبه نرم ضرب این ۲ ماتریس میزان همپوشانی این ۲ زیر فضا بدست می آید. در آزمایش ها این مقدار شباهت زیر فضا های ذخیره شده از بازنمایی ورودی ها برای لایه ها قبل و بعد از آموزش تسک ۲ بررسی می شود. همچنان این معیار در لایه های پایانی ضعیف تر از لایه های اولیه است. با توجه به پایین بودن معیار مشابهت فضای بدست آمده تسک اول و تسک اول پس از اموزش تسک دوم در لایه های آخر و بالا بودن مشابهت فضای تسک دوم و تسک اول پس از اموزش تسک اول با اینکه تسک های متفاوتی هستند، میبینیم که معیار پایین مرحله قبل ناشی از این است که آموزش تسک های بعدی فضای بازنمایی را در لایه های عمیق تر عوض کرده و پس از آموزش داده های تسک اول نیز در زیر فضای مناسب تسک دوم بازنمایی می شوند. به عبارتی این پاک کردن و یادگیری جدید زیر فضای بازنمایی در لایه های آخر باعث نمایش نامناسب فیچر های داده های تسک اول و فراموشی فضای بازنمایی مناسب تسک اول خواهد شد.
